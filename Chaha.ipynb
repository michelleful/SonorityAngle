```python
# Sonority scale employed for Chaha - note addition of the bilabial glide

SONORITY = {
    'T': 1.0,   # voiceless stops
    'F': 2.0,   # fricatives
    'N': 3.0,   # nasals
    'b': 3.5,   # Chaha bilabial glide
    'R': 4.0,   # liquids
    'G': 5.0,   # glides
    'V': 6.0,   # vowels
}
```

```python
# Chaha consonants

CONSONANT_CLASSES = {
    't': 'T', 'd': 'T', 'k': 'T', 'g': 'T',
    'f': 'F', 's': 'F', 'z': 'F', 'x': 'F',
    'n': 'N', 'm': 'N',
    'B': 'b', 
    'r': 'R',
    'w': 'G',
}
```

```python
# define the sonority angle metric
import math
def sonority_angle_class(C1, C2):
    return math.atan(SONORITY['V']-SONORITY[C1])-math.atan(SONORITY[C2]-SONORITY[C1])
def sonority_angle(C1, C2):
    return sonority_angle_class(CONSONANT_CLASSES[C1], CONSONANT_CLASSES[C2])
```

```python
# Resultant cluster hierarchy for Chaha
import operator
sonority_classes = ['T','F','N','b','R']
sonangles = {C1+C2: sonority_angle_class(C1,C2) for C1 in sonority_classes for C2 in sonority_classes}
sonangles['GF'] = sonority_angle_class('G','F')
for k, v in sorted(sonangles.items(), key=operator.itemgetter(1)):
    print k, "\t", v
```

CCC clusters
------------

Medial triconsonantal clusters are broken up by epenthesis either as [CCiC] or [CiCC]. There is considerable inter- and intra-speaker variation in this as well.

First we address the data presented in Section 4.3 of Rose 2000. In this idiolect, the majority of the clusters epenthesise as [CCiC]. Those that do not have a sonority fall between C2 and C3. If we had epenthesised between C2 and C3 in these cases, a large Sonority Angle would have been created, so the grammar then prefers to epenthesise between C1 and C2. In other words, a markedness constraint favours epenthesising between C2 and C3, but is dominated by Ident(SonAngle)$<$1.5, which is the cut-off between level and falling sonority clusters (between TT = 1.37 and RB = 1.57). All other Ident(SonAngle)$<$n constraints are ranked below markedness.

I adopt a markedness constraint used by Rose:

Align-Left: every consonant must be aligned with the left edge of some prosodic word.

This favours CCiC over CiCC.

When both Sonority Angles exceed 1.5, they both violate Ident(SonAngle)$<$1.5, and the markedness constraint preferring CCiC takes over - even if the Sonority Angle between C2 and C3 exceeds that of C1-C2. This is a consequence of adopting free ranking between the stringency hierarchy of Ident(SonAngle)$<$n constraints, rather than a universal ranking.

(d) Rising-Falling

When the Sonority Angle between C1 and C2 is less than 1.5 and the Sonority Angle between C2 and C3 is greater than 1.5, the faithfulness constraint enforces the marked outcome of CiCC.
    
Thus the default idiolect is explained. (According to Rose 2000 fn 22, /tBr/ is epenthesised as [tiBr], which is unexpected under both her and my account. I will ignore this fact.) The following two code cells verify that this ranking works.


```python
# List of medial 3-consonant clusters. 
# t', k' written as t, k respectively. Epenthetic vowel written as i. Schwa written as @. Bilabial sonorant written as B.
# On the left is the underlying cluster, on the right the list of possible outputs, the first one being the "default".

from collections import OrderedDict
CLUSTERS = OrderedDict([
   # Falling-Rising
   ('nkt', ['nkit']), # 24a. always CCiC
   ('nks', ['nkis']), # 24b. always CCiC
   ('nkm', ['nkim']), # 24c. always CCiC
   ('wzf', ['wzif', 'wizf']), # 24d. default (?) CCiC, also CiCC
   ('mxr', ['mxir']), # 24e. always CCiC
   ('sgd', ['sgid', 'sigd']), # 24f. default CCiC, also CiCC
   ('sdB', ['sdiB']), # 24g. always CCiC
   # Rising-Rising   
   ('gdf', ['gdif', 'gidf']), # 26a. default CCiC, also CiCC
   ('gdr', ['gdir', 'gidr']), # 26b. default CCiC, also CiCC
   ('kmr', ['kmir', 'kimr']), # 26c. default CCiC, also CiCC
   ('sfr', ['sfir', 'sifr']), # 26d. default CCiC, also CiCC
   ('sBr', ['sBir', 'siBr']), # 26e. default CCiC, also CiCC
   # Falling-Falling
   ('nfk', ['nfik']), # 28a. always CCiC
   ('rmd', ['rmid']), # 28b. always CCiC
   ('mst', ['msit', 'mist']), # 28c. Rose reports always CCiC; Degif Petros Banksira reports he can say j@misto (p.c.)
   # Rising-Falling
   ('kft', ['kift']), # 30a. always CiCC
   ('dmd', ['dimd']), # 30b. always CiCC
   ('drs', ['dirs']), # 30c. always CiCC
   ('krm', ['kirm', 'krim']), # 30d. default CiCC, also CCiC
   ('srB', ['sirB', 'sriB']), # 30e. default CiCC, also CCiC
   ('frt', ['firt']), # 30f. always CiCC
   ('sBx', ['siBx']), # 30g. always CiCC
   ('mrg', ['mirg']), # 30h. always CiCC
   # Additional non-jussive j-(t-CC)@C@C -> ji(tiCC/tCiC)@C@C
   ('tBt', ['tiBt']), # 32e. always CiCC; no variation reported
   ('trs', ['tirs']), # 32f. always CiCC; no variation reported
   ('tkB', ['tkiB']), # 32g. default CCiC, no variation indicated
   ('tkt', ['tkit', 'tikt']), # 32h. default CCiC, no variation indicated; but under Rose's second constraint ranking (51c) should also surface as CiCC
   # Footnote 22
   ('tBr', ['tiBr', 'tBir']), # reported always CCiC, unexpected under Rose's analysis
   ('tmr', ['tmir', 'timr']),
])

import itertools
PERMISSIBLE_OUTPUTS = set(itertools.chain(*CLUSTERS.values()))
```

```python
def align_cluster_output(cluster, n, jussive_r_on=True):
    """
    Returns output for cluster under the ranking Jussive-R, Ident(SonAngle)<n >> Align-L (favours CCiC)
    """
    if jussive_r_on:
        if cluster[0] in "rn":
            return cluster[0:2] + 'i' + cluster[2]
    # F >> M
    c1_c2_angle = sonority_angle(cluster[0],cluster[1])
    c2_c3_angle = sonority_angle(cluster[1],cluster[2])
    if c1_c2_angle < n and c2_c3_angle >= n: # F>>M
        return cluster[0] + 'i' + cluster[1:] # CiCC
    else: # M takes over
        return cluster[0:2] + 'i' + cluster[2] # CCiC

def align_behavior_for_n(n):
    print "Jussive-R, Ident(SonAngle)<%s >> Align-Left" % n
    for cluster in CLUSTERS.keys():
        output = align_cluster_output(cluster, n)
        permissible = "*" if output not in PERMISSIBLE_OUTPUTS else ""
        primary = "1" if output == CLUSTERS[cluster][0] else ""
        print cluster, "\t", output, "\t", permissible, "\t", primary    

# Default dialect
print "Verifying behaviour of default idiolect"
align_behavior_for_n(1.5)
```

In a second idiolect, clusters whose second consonant is /r/ and third consonant is a sonorant also epenthesise as [CCiC] rather than their default [CiCC].

The adjustment to this idiolect is quite straightforward: 

```python
# "Default dialect" but when C2=r, C3=son, epenthesise between C2 and C3

# Ident(SonAngle)<1.9 >> Align-L (favours CCiC)

align_behavior_for_n(1.9)
```

```python
# "CiCC dialect" where everything that can be CiCC, is due to this constraint ranking:

# Jussive-R, Ident(SonAngle)<n >> Anchor-L-Root (favours CCiC)

def anchor_cluster_output(cluster, n):
    if cluster[0] in 'rn': # effect of Jussive-R
        return cluster[0:2] + 'i' + cluster[2]    
    # the rest is the effect of Ident(SonAngle)<n >> Anchor-L-Root (favours CCiC)
    c1_c2_angle = sonority_angle(cluster[0],cluster[1])
    c2_c3_angle = sonority_angle(cluster[1],cluster[2])
    if c2_c3_angle < n and c1_c2_angle >= n:
        return cluster[0:2] + 'i' + cluster[2] # only return marked CCiC if C2-C3 angle < n AND C1-C2 angle >= n
    else:
        return cluster[0] + 'i' + cluster[1:] # default CiCC
    
def anchor_behavior_for_n(n):
    print "Jussive-R, Ident(SonAngle)<%s >> Anchor-L-Root" % n
    for cluster in CLUSTERS.keys():
        output = anchor_cluster_output(cluster, n)
        permissible = "*" if output not in PERMISSIBLE_OUTPUTS else ""
        cicc_possible = len(CLUSTERS[cluster]) == 2 or CLUSTERS[cluster][0][1]=="i"
        output_is_ccic_but_cicc_possible = ':(' if output[2]=="i" and cicc_possible else ""
        print cluster, "\t", output, "\t", permissible, "\t", output_is_ccic_but_cicc_possible
    
anchor_behavior_for_n(1.5)
print "\n\n"
anchor_behavior_for_n(1.0)
# Unfortunately, we need to use two different settings of n here to get all the CiCC clusters
    
```

```python
# CCCC cases - omitting those covered by Jussive-R, split geminate, etc

CCCC_CLUSTERS = OrderedDict([
    ('sdBn', ['sdiBn']), # 33b
    ('wzfn', ['wzifn', 'wizfin']), # 33c
    ('gdfn', ['gdifn', 'gidfin']), # 33d
    ('mstn', ['mistin']), # 34c
    ('dmdn', ['dimdin']), # 35a
    ('drsn', ['dirsin']), # 35b
    ('sBxn', ['siBxin']), # 35c
    ('kftn', ['kiftin']), # 35d
    ('ktfn', ['ktifn', 'kitfin']), # 44a
    ('sgdn', ['sgidn', 'sigdin']), # 48a
    ('srBn', ['sriBn', 'sirBin']), # 53a
    ('krmn', ['krimn', 'kirmin']), # 53b
])
# TODO: think about 53c, 53d

PERMISSIBLE_4C_OUTPUTS = set(itertools.chain(*CCCC_CLUSTERS.values()))
```

```python
# Dialect 1: 

# Jussive-R, Ident(SonAngle)<1.5 >> Align-L (favours CCiCC)

ALL_LEGAL_4C_OUTPUTS = set()

def align_cluster_output_4C(cluster, n):
    c1_c2_angle = sonority_angle(cluster[0],cluster[1])
    c2_c3_angle = sonority_angle(cluster[1],cluster[2])
    c3_c4_angle = sonority_angle(cluster[2],cluster[3])
    if c1_c2_angle < n and c3_c4_angle < n and c2_c3_angle >= n:
        return cluster[0] + 'i' + cluster[1:3] + 'i' + cluster[3]
    else:
        return cluster[0:2] + 'i' + cluster[2:]

def align_behavior_for_n_4C(n):
    print "Jussive-R, Ident(SonAngle)<%s >> Align-Left" % n
    for cluster in CCCC_CLUSTERS.keys():
        output = align_cluster_output_4C(cluster, n)
        permissible = "*" if output not in PERMISSIBLE_4C_OUTPUTS else ""
        if not permissible: # meaning it actually is - bad variable naming 
            ALL_LEGAL_4C_OUTPUTS.add(output)
        primary = "1" if output == CCCC_CLUSTERS[cluster][0] else ""
        print cluster, "\t", output, "\t", permissible, "\t", primary    

align_behavior_for_n_4C(1.5)
```

```python
# Dialect 3


def anchor_cluster_4C_output(cluster, n):
    c1_c2_angle = sonority_angle(cluster[0],cluster[1])
    c2_c3_angle = sonority_angle(cluster[1],cluster[2])
    c3_c4_angle = sonority_angle(cluster[2],cluster[3])
    if c2_c3_angle < n and (c1_c2_angle >= n or c3_c4_angle >=n):
        return cluster[0:2] + 'i' + cluster[2:]
    else:
        return cluster[0] + 'i' + cluster[1:3] + 'i' + cluster[3]
    
def anchor_behavior_for_n_4C(n):
    print "Jussive-R, Ident(SonAngle)<%s >> Anchor-L-Root" % n
    for cluster in CCCC_CLUSTERS.keys():
        output = anchor_cluster_4C_output(cluster, n)
        permissible = "*" if output not in PERMISSIBLE_4C_OUTPUTS else ""
        if not permissible: # meaning it actually is - bad variable naming 
            ALL_LEGAL_4C_OUTPUTS.add(output)
        print cluster, "\t", output, "\t", permissible
        
anchor_behavior_for_n_4C(1.5)
anchor_behavior_for_n_4C(1.3)
```
